```python
# 02 - Data Cleaning & Feature Engineering

This notebook prepares the dataset for modelling by:

- Cleaning inconsistent and missing data
- Filtering irrelevant observations
- Creating per-90 performance metrics
- Engineering meaningful football-specific features

The goal is to create a structured dataset suitable for machine learning models predicting player market value.
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../data/players_with_valuations.csv")

print("Shape:", df.shape)
df.head()
```

```python
# Remove duplicates
df = df.drop_duplicates()

print("After removing duplicates:", df.shape)
```

```python
# Minutes played
df["Min"] = df["Min"].astype(str).str.replace(",", "")
df["Min"] = pd.to_numeric(df["Min"], errors="coerce")

# Age
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")

# 90s
df["90s"] = pd.to_numeric(df["90s"], errors="coerce")
```

```python
# Remove low-minute players
df = df[df["Min"] >= 500]

print("After filtering low-minute players:", df.shape)
```

```python
missing = df.isnull().sum().sort_values(ascending=False)
missing.head(20)
```

```python
# Simple strategy (safe for now)
df = df.fillna(0)
```

```python
stat_cols = ["Gls", "Ast", "Sh", "SoT"]

for col in stat_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
for col in stat_cols:
    if col in df.columns:
        df[f"{col}_per90"] = df[col] / df["90s"]
```

```python
# Goal contribution
df["goal_contribution"] = df["Gls"] + df["Ast"]

# Goal contribution per 90
df["goal_contribution_per90"] = df["goal_contribution"] / df["90s"]

# Shooting accuracy
df["shot_accuracy"] = df["SoT"] / df["Sh"]
```

```python
# Simple attacking score
df["attacking_score"] = (
    df["Gls_per90"] * 0.5 +
    df["Ast_per90"] * 0.3 +
    df["Sh_per90"] * 0.2
)
```

```python
def simplify_position(pos):
    if "GK" in pos:
        return "GK"
    elif "DF" in pos:
        return "DEF"
    elif "MF" in pos:
        return "MID"
    elif "FW" in pos:
        return "FWD"
    else:
        return "OTHER"

df["position_group"] = df["Pos"].apply(simplify_position)
```

```python
df = pd.get_dummies(df, columns=["Comp"], drop_first=False)
```

```python
print("Final dataset shape:", df.shape)
df.head()
```

```python
df.to_csv("../data/cleaned_players_dataset.csv", index=False)

print("Cleaned dataset saved.")
```
